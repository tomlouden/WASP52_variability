# Lomb scargle periodograms

```python
import trm.pgram as pgram

ofac  = 2.
hifac = 1.
f,p = pgram.fwmls(wasp_dict['HJD'],wasp_dict['mag'],wasp_dict['err'],ofac,hifac)
```

```python
ax = plt.subplot(111)

ax.plot(1.0/f,p,'k')
ax.set_xlim(0.4,400)
ax.set_xscale('log')

plt.show()
```

Looks interesting, strong periodicity at ~ 19 days. How does this compare to previous analysis of data? G. Hebrard et al analyse the two seasons seperately - we should try that too, particularly since we have a new season now which may be done with a different camera

```python
s1 = wasp_dict['HJD'] < 4900
s2 = (wasp_dict['HJD'] > 4900) & (wasp_dict['HJD'] < 5300)
s3 = wasp_dict['HJD'] > 5300

f1,p1 = pgram.fwmls(wasp_dict['HJD'][s1],wasp_dict['mag'][s1],wasp_dict['err'][s1],ofac,hifac)
f2,p2 = pgram.fwmls(wasp_dict['HJD'][s2],wasp_dict['mag'][s2],wasp_dict['err'][s2],ofac,hifac)
f3,p3 = pgram.fwmls(wasp_dict['HJD'][s3],wasp_dict['mag'][s3],wasp_dict['err'][s3],ofac,hifac)
```

```python
fig, axarr = plt.subplots(2,2)

fig.set_size_inches(16, 9)

axarr[0,0].plot(1.0/f1,p1,'k')
axarr[0,0].set_xlim(0.4,400)
axarr[0,0].set_xscale('log')
axarr[0,0].set_title('Season 1')
axarr[0,0].set_ylabel('Power')

axarr[0,1].plot(1.0/f2,p2,'k')
axarr[0,1].set_xlim(0.4,400)
axarr[0,1].set_xscale('log')
axarr[0,1].set_title('Season 2')
axarr[0,1].set_ylabel('Power')

axarr[1,0].plot(1.0/f3,p3,'k')
axarr[1,0].set_xlim(0.4,400)
axarr[1,0].set_xscale('log')
axarr[1,0].set_title('Season 3')
axarr[1,0].set_xlabel('Frequency (days)')
axarr[1,0].set_ylabel('Power')

axarr[1,1].plot(1.0/f,p,'k')
axarr[1,1].set_xlim(0.4,400)
axarr[1,1].set_xscale('log')
axarr[1,1].set_title('All seasons')
axarr[1,1].set_xlabel('Frequency (days)')
axarr[1,1].set_ylabel('Power')

plt.savefig('figures/lomb_scargles.pdf')

plt.show()
```

It's not at all obvious what's going on here, and whether it's useful. I think it probably needs some binning?
Periodicity was immediately obvious in Lomb Scargle. There's a 1 day peak at least, which indicates that this isn't just noise...

```python
bins = 100

print len(result)

slices = np.linspace(0, len(result), len(result)/bins, True).astype(np.int)

print len(slices)

counts = np.diff(slices)

bin_result = np.add.reduceat(result, slices[:-1]) / counts
bin_periods = np.add.reduceat(periods, slices[:-1]) / counts
```

```python
fig, ax1 = plt.subplots(1,1)

fig.set_size_inches(16, 9)

ax1.set_xlim(0.4,400)
ax1.set_ylim(-0.05,0.05)
ax1.set_xscale('log')
ax1.set_title('Season 3')
ax1.set_xlabel('Frequency (days)')
ax1.set_ylabel('Power')

ax1.plot(bin_periods,bin_result,'k',linewidth=0.5)

plt.show()
```

Figured it out: The reason this looks like shit is because we need a WEIGHTED autocorrelation function. All of those 0 points are being counted as bad correlations. padding out with zeros is clearly wrong...

need a new technique - discrete correlation function Edelson 1988

```python
from WASP_lc import *
import matplotlib.pyplot as plt
import pickle
%pylab inline

nproc = 16

fname = '1SWASPJ231358.76+084540.6_WASP52B_100_ORKP_TAMTFA.lc'
wasp_dict = load_lc('lightcurves/'+fname)

# clip out bad values (actually it turns out this was only one value but it was fucking up everything.
good_vals = [(wasp_dict['mag'] != 0.0) & (wasp_dict['err'] != 0.0)]
wasp_dict['HJD'] = wasp_dict['HJD'][good_vals]
wasp_dict['mag'] = wasp_dict['mag'][good_vals]
wasp_dict['err'] = wasp_dict['err'][good_vals]

s1 = wasp_dict['HJD'] < 4900
s2 = (wasp_dict['HJD'] > 4900) & (wasp_dict['HJD'] < 5200)
s3 = wasp_dict['HJD'] > 5200

HJD_s1 = wasp_dict['HJD'][s1]
mag_s1 = wasp_dict['mag'][s1]
err_s1 = wasp_dict['err'][s1]
HJD_s1, mag_s1, err_s1 = sum_duplicates(HJD_s1,mag_s1,err_s1)

HJD_s2 = wasp_dict['HJD'][s2]
mag_s2 = wasp_dict['mag'][s2]
err_s2 = wasp_dict['err'][s2]
HJD_s2, mag_s2, err_s2 = sum_duplicates(HJD_s2,mag_s2,err_s2)

HJD_s3 = wasp_dict['HJD'][s3]
mag_s3 = wasp_dict['mag'][s3]
err_s3 = wasp_dict['err'][s3]
HJD_s3, mag_s3, err_s3 = sum_duplicates(HJD_s3,mag_s3,err_s3)
```
